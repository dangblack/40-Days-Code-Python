```python
# Import thư viện cần thiết
import pandas as pd
import sqlite3
```

```python
# Kết nối đến cơ sở dữ liệu SQLite
connection = sqlite3.connect('database.sqlite')

```

```python
# Tạo con trỏ truy vấn
cursor = connection.cursor()

# Thực thi truy vấn để lấy tổng giá trị mua
data_sum = pd.read_sql_query("select sum(buy) from STOCK", connection)

# In ra tổng giá trị mua
print(data_sum)
```

```python

# Thực thi truy vấn để lấy thông tin chi tiết về các cổ phiếu
cursor.execute("""
SELECT investor, MAX(buy) AS MAX_PRICE
FROM STOCK
GROUP BY investor
ORDER BY MAX_PRICE DESC
""")

```

```python
# Lấy kết quả truy vấn
data_detail = cursor.fetchall()

# In ra kết quả truy vấn
for row in data_detail:
    print(f"Investor: {row[0]}")
    print(f"MAX_PRICE: {row[1]}")
```

```python
# Thêm dữ liệu mới vào bảng STOCK
cursor.execute("""
INSERT INTO STOCK (id, name, buy, investor) VALUES
(1, 'ACB', 29.45, 'Nam'),
(2, 'VIC', 44.55, 'Nam'),
(3, 'GMD', 74.30, 'Nam'),
(4, 'ACB', 28.45, 'An'),
(5, 'VIC', 40.55, 'An'),
(6, 'GMD', 60.30, 'An')
""")

# Lưu thay đổi vào cơ sở dữ liệu
connection.commit()

# Đóng kết nối đến cơ sở dữ liệu
connection.close()
```
