```python
import sqlite3
import pandas as pd

```

```python
# Kết nối đến cơ sở dữ liệu SQLite
con = sqlite3.connect('sqlite.db')
cur = con.cursor()

```

```python
# Tạo bảng PRODUCT
cur.execute("""
CREATE TABLE PRODUCT (
    ID INTEGER PRIMARY KEY,
    NAME TEXT NOT NULL,
    PRICE INTEGER NOT NULL
);
""")

```

```python
# Thêm dữ liệu vào bảng PRODUCT
cur.execute("""
INSERT INTO PRODUCT (ID, NAME, PRICE)
VALUES
    ('1', 'iPhone 15', '18000000'),
    ('2', 'Galaxy Z-Fold 5', '30000000')
""")

# Ghi lại thay đổi
con.commit()

```

```python
# Đọc dữ liệu từ bảng PRODUCT
data = pd.read_sql_query("SELECT * FROM PRODUCT", con)
print(data)

```

```python
# Cập nhật giá của Galaxy Z-Fold 5
cur.execute("""
UPDATE PRODUCT
SET PRICE = '50000000'
WHERE NAME = 'Galaxy Z-Fold 5'
""")

# Ghi lại thay đổi
con.commit()

```

```python
# Đọc dữ liệu từ bảng PRODUCT
data = pd.read_sql_query("SELECT * FROM PRODUCT", con)
print(data)

```

```python
# Xóa iPhone 15 khỏi bảng PRODUCT
cur.execute("""
DELETE FROM PRODUCT
WHERE NAME = 'iPhone 15'
""")

# Ghi lại thay đổi
con.commit()

```

```python
# Đọc dữ liệu từ bảng PRODUCT
data = pd.read_sql_query("SELECT * FROM PRODUCT", con)
print(data)
```
