```python
from IPython.display import clear_output
import random
total_money = 1000000
```

```python
def generate_lottery_results():
    lottery_results = []

    for _ in range(27):
        result = ""
        for _ in range(7):
            digit = random.randint(0, 9)  # Tạo một chữ số ngẫu nhiên từ 0 đến 9
            result += str(digit)  # Chuyển chữ số thành chuỗi và thêm vào kết quả
        lottery_results.append(result)

    return lottery_results

def lottery_cal(choices, bet_amount, lst_result):
    global total_money
    count = 0
    money_loss = 0

    for choice in choices:
        if len(choice) != 2 or not choice.isdigit():
            return "Lô bạn nhập không hợp lệ. Vui lòng chọn các số có 2 chữ số từ 00 đến 99."

    for choice in choices:
        choice_matched = False
        for result in lst_result:
            if result[-2:] == choice:
                count += 1
                choice_matched = True
                break  # Thoát khỏi vòng lặp khi đã trúng lô

        if not choice_matched:
            money_loss += bet_amount  # Tính tiền cược bị mất cho mỗi lựa chọn không trúng

    print("Danh sách quay số: \n",lst_result)
    if count > 0:
        print(f"Bạn đã trúng {count} con lô")
        total_money += bet_amount * 70 * count - money_loss
    else:
        print("Bạn đã tạch lô")

    return total_money

def lottery_game():
    clear_output()
    while True:
        choices = input("Nhập các số lô bạn muốn chơi, cách nhau bằng dấu phẩy: ")
        lst_choices = choices.split(",")

        # Kiểm tra xem các lựa chọn có hợp lệ không
        valid_choices = True
        for choice in lst_choices:
            if len(choice) != 2 or not choice.isdigit():
                valid_choices = False
                break

        if not valid_choices:
            print("Một số bạn nhập con lô không hợp lệ. Vui lòng nhập lại.")
            continue

        break

    bet_amount = int(input("Nhập số tiền cược/con lô (tổng tiền sẽ nhân với số lô mà bạn cược): "))
    total_money = lottery_cal(lst_choices, bet_amount, generate_lottery_results())
    print("Tiền của bạn sau khi chơi lô là: ", total_money)
def mini_lottery_game():
    clear_output(wait=True)
    while True:
        lottery_game()
        option = input("Bạn có muốn chơi tiếp không? (0: Không - quay về Menu, 1: Có): ")
        if option == "1":
            continue
        else:
            main()  # Quay lại menu chính
            break
```

```python
def recharge_money():
    global total_money  # Sử dụng total_money như một biến toàn cục
    print("Số tiền bạn đang có là: ", total_money)
    while True:
        recharge = int(input("Nhập số tiền mà bạn muốn nạp: "))
        total_money += recharge
        print("Số tiền sau khi bạn nạp vào là: ", total_money)
        option = input("Bạn có muốn nạp thêm không? (0: Không, 1: Có): ")
        if option == "1":
            continue
        else:
            main()  # Quay lại menu chính
            break

```

```python
def exit_game():
    print("Thoát game thành công. Hẹn gặp lại bạn :)")
    return
```

```python
def switch_option(option):
    clear_output()
    options = {
        1: mini_lottery_game,
        2: recharge_money,
        3: exit_game
    }
    try:
        clear_output(wait=True)
        options[option]()
    except KeyError:
        print("Có lỗi xảy ra trong quá trình thực hiện switch case")

def menu():
    clear_output()
    print("------- GAME LÔ ĐỀ HỌC --------")
    print("Bạn vui lòng chọn 1 lựa chọn")
    print("1. Chơi game Lô đề")
    print("2. Nạp tiền chơi game")
    print("3. Thoát game")

def main():
    while True:
        menu()
        try:
            option = int(input("Lựa chọn của bạn: "))
            if option not in [1, 2, 3]:
                raise ValueError("Lựa chọn không hợp lệ, vui lòng nhập lại")
            switch_option(option)
            break
        except ValueError as e:
            print(e)
```
