```python
matrix_A = [
    [1,2,3],
    [4,5,6],
    [7,8,9]
]

matrix_B = [
    [2,4,6],
    [1,3,5],
    [1,0,1]
]
```

```python
sum = []
sub = []
for i in range(len(matrix_A)):
  _sum = []
  _sub = []
  for j in range(len(matrix_B)):
    _sum.append(matrix_A[i][j] + matrix_B[i][j])
    _sub.append(matrix_A[i][j] - matrix_B[i][j])
  sum.append(_sum)
  sub.append(_sub)
```

```python
print(f"Sum matrix A + B= {sum}")
print(f"Sub matrix A - B= {sub}")
```

```python
dot = []
for i in range(len(matrix_A)):
  _dots = []
  for j in range(len(matrix_B)):
    _dot = 0
    for k in range(len(matrix_A)):
      _dot += matrix_A[i][k] * matrix_B[k][j]
    _dots.append(_dot)
  dot.append(_dots)
print(f"Dot matrix A * B= {dot}")
```
