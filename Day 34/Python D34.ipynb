```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('/content/drive/MyDrive/data/data set/2.jpg', 0)
plt.imshow(img)
plt.title('Original')
plt.show()

img_filtered_1 = cv2.GaussianBlur(img, (5,5), 2.5)
plt.imshow(img_filtered_1)
plt.title('Sigma=2.5')
plt.show()

img_filtered_2 = cv2.GaussianBlur(img, (5,5), 5.0)
plt.imshow(img_filtered_2)
plt.title('Sigma=5.0')
plt.show()

img_filtered_3 = cv2.GaussianBlur(img, (5,5), 10.0)
plt.imshow(img_filtered_3)
plt.title('Sigma=10.0')
plt.show()
```
