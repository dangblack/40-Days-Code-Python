```python
with open('/content/drive/MyDrive/Study/AI VIETNAM/Python Practice/Day 21/data.txt', 'r') as f:
    data = f.read()
data
```

```python
def remove_characters(input, characters):
    for char in characters:
        input = input.replace(char, ' ')
        return input

data = remove_characters(data, '\n')
```

```python
# Remove \n and lower character
data = data.replace('\n', ' ').lower()
```

```python
data
```

```python
# find distinct words


distinict_words=set(data.split())

distinict_words
```

```python
distinict_words
```

```python
words = data.split(' ')

word_counts = {}

for word in words:
    if word in word_counts:
        word_counts[word] += 1
    else:
        word_counts[word] = 1

# Tìm từ xuất hiện nhiều nhất và ít nhất
most_common_word = max(word_counts, key=word_counts.get)
least_common_word = min(word_counts, key=word_counts.get)

# Tìm số lần xuất hiện của từ xuất hiện nhiều nhất và ít nhất
most_common_count = word_counts[most_common_word]
least_common_count = word_counts[least_common_word]

print("Từ xuất hiện nhiều nhất là '{}' với {} lần xuất hiện.".format(most_common_word, most_common_count))
print("Từ xuất hiện ít nhất là '{}' với {} lần xuất hiện.".format(least_common_word, least_common_count))
```
