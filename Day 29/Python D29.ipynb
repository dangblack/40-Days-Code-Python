```python
import librosa

path = "/content/drive/MyDrive/data/Chúng-Ta-Của-Tương-Lai.wav"
audio, sr = librosa.load(path)

import numpy as np
import matplotlib.pyplot as plt
time = np.linspace(0, audio.shape[0]/sr, num=audio.shape[0])
plt.plot(time, audio, color="blue")
plt.xlabel("Time (s)")
plt.xlabel("Amplitude (quantized)")
plt.title("Wav file visualization")
plt.axhline(y=0, color="r", linestyle="-")
plt.show()
```

```python
import speech_recognition as sr

r = sr.Recognizer()
audio_filename = "/content/drive/MyDrive/data/Chúng-Ta-Của-Tương-Lai.wav"

my_audio = sr.AudioFile(audio_filename)
with my_audio as source:
    audio = r.record(source)

print(type(audio))
your_speech = r.recognize_google(audio, language="vi-VN")
print("Audio transcription: ", your_speech)
```

```python
from gtts import gTTS
import librosa
import numpy as np
import IPython.display

lang = "vi"
output_filename = "/content/drive/MyDrive/data/record.mp3"
content = "Tôi yêu AIO"
output = gTTS(content, lang=lang, slow=False)

output.save(output_filename)
data, sr = librosa.load(output_filename)

IPython.display.display(IPython.display.Audio(np.transpose(data), rate=sr))
```
