```python
from pypdf import PdfReader

reader = PdfReader("/2402.13616.pdf")
count = 0

for page in reader.pages :
  for image_file_object in page.images:
    with open (str(count) + image_file_object.name , "wb") as fp :
      fp.write(image_file_object.data)
      count += 1
```

```python
# Merge PDFs
from pypdf import PdfWriter

merger = PdfWriter()
for pdf in ["/content/drive/MyDrive/Study/AI VIETNAM/Python Practice/Day 37/2402.13616.pdf", "/content/drive/MyDrive/Study/AI VIETNAM/Python Practice/Day 37/day-37.pdf"]:
  merger.append (pdf)
  merger.write("merged-yolov-python-practive.pdf")
  merger.close()
```

```python
from pypdf import PdfReader
# Đọc file PDF
reader = PdfReader("/content/merged-yolov-python-practive.pdf")

num_pages = len(reader.pages)
page_2 = reader.pages [0]
page_2_txt = page_2.extract_text ()
pages_txt = ""

for i in range (num_pages):
  page = reader.pages[i]
  pages_txt += page.extract_text () + "\n"

print(pages_txt)
```

```python
from pypdf import PdfWriter

writer = PdfWriter (clone_from ="/content/merged-yolov-python-practive.pdf")

for page in writer.pages :
  page.compress_content_streams(level =8)

with open ("out.pdf ", "wb") as f:
  writer.write(f)
```
